```python
#MODA
def moda(datos):
    frecuencias = {}

    for c in datos:
        if c in frecuencias:
            frecuencias[c] += 1
        else:
            frecuencias[c] = 1

    mayor_frecuencia = 0
    moda_valor = None

    for valor in frecuencias:
        if frecuencias[valor] > mayor_frecuencia:
            mayor_frecuencia = frecuencias[valor]
            moda_valor = valor

    return moda_valor
```

```python
#MEDIANA
def mediana(lista):
    lista = sorted(lista)
    f = len(lista)
    e = f // 2
    if f % 2 == 0:
        return (lista[e-1] + lista[e]) / 2
    else:
        return lista[e]
```

```python
#PROMEDIO
def promedio(lista):
    total = 0
    for c in lista:
        total += c
    return total / len(lista)
```

```python
#DESVIACIÓN ESTANDAR
def desviacion_estandar(lista):
    prom = promedio(lista)
    suma = 0
    for c in lista:
        suma += (c - prom) ** 2
    return (suma / len(lista)) ** 0.5
```

```python
#CUARTIL
def cuartiles(lista):
    lista = sorted(lista)
    f = len(lista)
    e = f // 2
    if f % 2 == 0:
        Q1 = mediana(lista[:e])
        Q3 = mediana(lista[e:])
    else:
        Q1 = mediana(lista[:e])
        Q3 = mediana(lista[e+1:])
    return Q1, Q3
```

```python
#PERCENTIL
def percentil(lista, p):
    lista = sorted(lista)
    a = (len(lista)-1) * p / 100
    b = int(a)
    c = b + 1
    if c >= len(lista):
        return lista[b]
    return lista[b] + (a-b)*(lista[c]-lista[b])
```

```python
#DESVIACIÓN MEDIA ABSOLUTA
def MAD(lista):
    med = mediana(lista)
    desv = []
    for c in lista:
        desv.append(abs(c - med))
    return mediana(desv)
```
