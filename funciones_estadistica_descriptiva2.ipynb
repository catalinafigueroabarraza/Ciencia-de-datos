```python
#DESVIACIÓN MEDIA ABSOLUTA
def MAD(lista):
    med = mediana(lista)
    desv = []
    for c in lista:
        desv.append(abs(c - med))
    return mediana(desv)
```

```python
#COVARIANZA
def covarianza(x, y):
    prom_x = promedio(x)
    prom_y = promedio(y)
    suma = 0
    for i in range(len(x)):
        suma += (x[i] - prom_x)*(y[i] - prom_y)
    return suma / len(x)
```

```python
#CORRELACIÓN
def correlacion(x, y):
    return covarianza(x, y) / (desviacion_estandar(x)*desviacion_estandar(y))
```
